### **1.Consumo di alcol per paese**
Esegui un’**analisi esplorativa** focalizzata sul **consumo di alcol**.  
Analizza i seguenti aspetti:

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/fivethirtyeight/data/master/alcohol-consumption/drinks.csv")
df
```

```python
df.sort_values(by="total_litres_of_pure_alcohol", ascending=False, inplace=True)
print(df.head(10))
import matplotlib.pyplot as plt

plt.bar(height=df.total_litres_of_pure_alcohol.head(10),x=df.country.head(10))
plt.xticks(rotation=45, ha='right')
```

```python
df.beer_servings.mean()
df["alcohol_index"]=(df.beer_servings + df.wine_servings + df.spirit_servings) / 3
df,df.alcohol_index.max()
```

```python
df[df.beer_servings>100]
```

```python
df1=df.sort_values(by="country").head(20)
plt.plot(df1.wine_servings)
plt.title("Wine Servings")
plt.xticks(ticks=range(len(df1.country)),labels=df1.country,rotation=45, ha='right')
 
```

### **2.Analisi offerte di lavoro**

Esegui un’**analisi esplorativa** focalizzata sui **lavori nel campo dei dati in ciascun paese**.  
Analizza i seguenti aspetti per ogni paese:

* Lo **stipendio medio annuale** (`salary_year_avg`)
* Il **numero totale di offerte di lavoro** (job count)
* L’intervallo degli stipendi: **valore minimo e massimo** (`min` / `max`)

Rappresenta graficamente lo stipendio medio (`salary_year_avg`) per `job_title_short` usando un grafico a barre orizzontali. Ordina i dati dallo stipendio più alto al più basso. Includi un titolo e le etichette per gli assi x e y.

```python
import pandas as pd

from datasets import load_dataset

dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df
```

```python
import numpy as np

df1=df.dropna(subset=["salary_year_avg"], )
df1.groupby("job_country")["salary_year_avg"].mean()
```

```python
df.job_country.value_counts()
```

```python
df1.groupby("job_country")["salary_year_avg"].agg(["min","max"])
```

```python
sal_m_jts = df1.groupby("job_title_short")["salary_year_avg"].mean()
sal_m_jts = sal_m_jts.sort_values(ascending=False)

plt.barh(sal_m_jts.index, sal_m_jts.values)
plt.xlabel("Stipendio medio annuale")
plt.ylabel("Job Title Short")
plt.title("Stipendio medio annuale per job_title_short")
plt.show()
```
